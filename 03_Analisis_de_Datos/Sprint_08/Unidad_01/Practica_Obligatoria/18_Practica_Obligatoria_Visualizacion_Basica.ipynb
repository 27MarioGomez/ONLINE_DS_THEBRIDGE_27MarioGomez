### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Ordenar las ciudades por población total
top_cities = df.sort_values(by="population_total", ascending=False).head(5)

# Crear un gráfico de barras de las ciudades más pobladas
plt.bar(top_cities["city"], top_cities["population_total"], color=["blue", "gray", "green", "red", "purple"])
plt.title("Ciudades más pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Número de habitantes")
plt.show()
```

```python
# Histograma para la distribución de la superficie
plt.hist(df["area_total_km2"], bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Distribución de la superficie (km²)")
plt.xlabel("Superficie (km²)")
plt.ylabel("Número de ciudades")
plt.show()
```

```python
# Scatter plot de densidad de población vs altitud
plt.scatter(df["elevation_m"], df["population_total"] / df["area_land_km2"], color="green", alpha=0.6)
plt.title("Densidad de población por kilómetro cuadrado y altitud")
plt.xlabel("Altitud (m)")
plt.ylabel("Densidad de población (hab/km²)")
plt.show()
```

```python
# Boxplot de la altitud
sns.boxplot(data=df, x="elevation_m", color="red")
plt.title("Diagrama de caja de la altitud")
plt.xlabel("Altitud (m)")
plt.show()
```
