### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)

3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Ordenar las ciudades por poblacin total
top_cities = df.sort_values(by="population_total", ascending=False).head(5)

# Crear un grfico de barras de las ciudades ms pobladas
plt.bar(top_cities["city"], top_cities["population_total"], color=["blue", "gray", "green", "red", "purple"])
plt.title("Ciudades ms pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Nmero de habitantes")
plt.show()
```

```python
# Histograma para la distribucin de la superficie
plt.hist(df["area_total_km2"], bins=30, color="skyblue", edgecolor="black", alpha=0.7)
plt.title("Distribucin de la superficie (km)")
plt.xlabel("Superficie (km)")
plt.ylabel("Nmero de ciudades")
plt.show()
```

```python
# Scatter plot de densidad de poblacin vs altitud
plt.scatter(df["elevation_m"], df["population_total"] / df["area_land_km2"], color="green", alpha=0.6)
plt.title("Densidad de poblacin por kilmetro cuadrado y altitud")
plt.xlabel("Altitud (m)")
plt.ylabel("Densidad de poblacin (hab/km)")
plt.show()
```

```python
# Boxplot de la altitud
sns.boxplot(data=df, x="elevation_m", color="red")
plt.title("Diagrama de caja de la altitud")
plt.xlabel("Altitud (m)")
plt.show()
```
