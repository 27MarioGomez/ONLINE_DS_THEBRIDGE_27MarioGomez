### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def scatter_plot(df, x_col, y_col, color_col=None, size_col=None, scale=1, show_legend=False):
    """
    Funcin para pintar diagramas de dispersin.
    
    Args:
    - df (pd.DataFrame): DataFrame con los datos.
    - x_col (str): Columna para el eje X.
    - y_col (str): Columna para el eje Y.
    - color_col (str, opcional): Columna categrica para definir los colores.
    - size_col (str, opcional): Columna para definir el tamao de los puntos.
    - scale (float): Factor para escalar el tamao de los puntos.
    - show_legend (bool): Si True, muestra una leyenda o un colorbar.
    """
    plt.figure(figsize=(12, 8))
    
    # Determinar el tamao de los puntos
    if size_col:
        sizes = df[size_col] * scale
    else:
        sizes = 50  # Tamao por defecto

    # Crear el grfico de dispersin
    if color_col:
        scatter = plt.scatter(df[x_col], df[y_col], c=df[color_col], s=sizes, alpha=0.6, cmap="viridis")
        if show_legend:
            plt.colorbar(scatter, label=color_col)
    else:
        plt.scatter(df[x_col], df[y_col], s=sizes, alpha=0.6)
    
    # Configurar etiquetas y ttulo
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f"Diagrama de dispersin: {x_col} vs {y_col}")
    
    if show_legend and not color_col:
        plt.legend(title="Tamao")

    plt.grid(True)
    plt.show()
```

```python
# Configuracin general
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Cargar el dataset
titanic_path = r"C:\Users\mario\Documents\GitHub\ONLINE_DS_THEBRIDGE_27MarioGomez\03_Analisis_de_Datos\Sprint_08\Unidad_02\Practica_Obligatoria\data\titanic.csv"
df_titanic = pd.read_csv(titanic_path)

# Vista previa del dataset
print(df_titanic.head())
```

```python
# Calcular tasa de supervivencia
survival_rates = df_titanic["survived"].value_counts(normalize=True) * 100

# Preparar datos para el grfico
categories = ["Murieron", "Sobrevivieron"]
values = survival_rates.values

# Crear el grfico de barras
plt.bar(categories, values, color=["red", "green"])  # Asignar colores directamente
plt.title("Supervivencia vs Mortalidad en el Titanic")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Estado")
plt.show()
```

```python
# Calcular las tasas de supervivencia para mujeres y nios
mujeres_nios_data = df_titanic.groupby("mujeres_nios")["survived"].mean() * 100
categories = ["Otros", "Mujeres y Nios"]
values = mujeres_nios_data.values

# Crear el grfico de barras
plt.bar(categories, values, color=["lightblue", "pink"])  # Colores directos
plt.title("Supervivencia: Mujeres y Nios Primero")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Grupo")
plt.show()
```

```python
# Calcular las tasas de supervivencia por clase
class_data = df_titanic.groupby("pclass")["survived"].mean() * 100
categories = ["Primera", "Segunda", "Tercera"]
values = class_data.values

# Crear el grfico de barras
plt.bar(categories, values, color=["darkblue", "blue", "lightblue"])  # Colores directos
plt.title("Tasa de Supervivencia segn Clase del Pasaje")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Clase del Pasaje")
plt.show()
```

```python
# Preparar datos para el grfico
class_sex_data = df_titanic.groupby(["pclass", "sex"])["survived"].mean().unstack() * 100
categories = ["Primera", "Segunda", "Tercera"]
male_values = class_sex_data["male"].values
female_values = class_sex_data["female"].values

# Crear el grfico de barras apiladas
x = range(len(categories))
width = 0.4

plt.bar(x, male_values, width=width, label="Hombres", color="blue", align="center")
plt.bar(x, female_values, width=width, label="Mujeres", color="pink", align="edge")
plt.xticks(x, categories)
plt.title("Supervivencia por Clase y Sexo")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Clase del Pasaje")
plt.legend()
plt.show()
```

```python
# Preparar datos para el grfico
embarked_class_data = df_titanic.groupby(["embarked", "pclass"])["survived"].mean().unstack() * 100
ports = ["Cherburgo", "Queenstown", "Southampton"]
first_class = embarked_class_data[1].values
second_class = embarked_class_data[2].values
third_class = embarked_class_data[3].values

# Crear el grfico de barras agrupadas
x = range(len(ports))
width = 0.2

plt.bar([i - width for i in x], first_class, width=width, label="Primera Clase", color="darkblue")
plt.bar(x, second_class, width=width, label="Segunda Clase", color="blue")
plt.bar([i + width for i in x], third_class, width=width, label="Tercera Clase", color="lightblue")
plt.xticks(x, ports)
plt.title("Supervivencia segn Puerto de Embarque y Clase del Pasaje")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Puerto de Embarque")
plt.legend()
plt.show()
```

```python
import os

# Crear carpeta para guardar las grficas
output_dir = "Titanic_Graphics"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Grfico 1: Tasa de Mortalidad
survival_rates = df_titanic["survived"].value_counts(normalize=True) * 100
categories = ["Murieron", "Sobrevivieron"]
values = survival_rates.values

plt.figure()
plt.bar(categories, values, color=["red", "green"])
plt.title("Supervivencia vs Mortalidad en el Titanic")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Estado")
plt.savefig(os.path.join(output_dir, "tasa_mortalidad.png"))
plt.close()

# Grfico 2: Mujeres y nios primero
mujeres_nios_data = df_titanic.groupby("mujeres_nios")["survived"].mean() * 100
categories = ["Otros", "Mujeres y Nios"]
values = mujeres_nios_data.values

plt.figure()
plt.bar(categories, values, color=["lightblue", "pink"])
plt.title("Supervivencia: Mujeres y Nios Primero")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Grupo")
plt.savefig(os.path.join(output_dir, "mujeres_nios.png"))
plt.close()

# Grfico 3: Influencia de la clase
class_data = df_titanic.groupby("pclass")["survived"].mean() * 100
categories = ["Primera", "Segunda", "Tercera"]
values = class_data.values

plt.figure()
plt.bar(categories, values, color=["darkblue", "blue", "lightblue"])
plt.title("Tasa de Supervivencia segn Clase del Pasaje")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Clase del Pasaje")
plt.savefig(os.path.join(output_dir, "influencia_clase.png"))
plt.close()

# Grfico 4: Relacin entre clase y gnero
class_sex_data = df_titanic.groupby(["pclass", "sex"])["survived"].mean().unstack() * 100
categories = ["Primera", "Segunda", "Tercera"]
male_values = class_sex_data["male"].values
female_values = class_sex_data["female"].values
x = range(len(categories))
width = 0.4

plt.figure()
plt.bar(x, male_values, width=width, label="Hombres", color="blue", align="center")
plt.bar(x, female_values, width=width, label="Mujeres", color="pink", align="edge")
plt.xticks(x, categories)
plt.title("Supervivencia por Clase y Sexo")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Clase del Pasaje")
plt.legend()
plt.savefig(os.path.join(output_dir, "clase_genero.png"))
plt.close()

# Grfico 5: Puerto de embarque y tipo de pasaje
embarked_class_data = df_titanic.groupby(["embarked", "pclass"])["survived"].mean().unstack() * 100
ports = ["Cherburgo", "Queenstown", "Southampton"]
first_class = embarked_class_data[1].values
second_class = embarked_class_data[2].values
third_class = embarked_class_data[3].values
x = range(len(ports))
width = 0.2

plt.figure()
plt.bar([i - width for i in x], first_class, width=width, label="Primera Clase", color="darkblue")
plt.bar(x, second_class, width=width, label="Segunda Clase", color="blue")
plt.bar([i + width for i in x], third_class, width=width, label="Tercera Clase", color="lightblue")
plt.xticks(x, ports)
plt.title("Supervivencia segn Puerto de Embarque y Clase del Pasaje")
plt.ylabel("Tasa de Supervivencia (%)")
plt.xlabel("Puerto de Embarque")
plt.legend()
plt.savefig(os.path.join(output_dir, "embarcacion_clase.png"))
plt.close()

print(f"Grficas guardadas en la carpeta: {output_dir}")
```
